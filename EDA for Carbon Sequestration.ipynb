```python
#Here we map soil types by FAO classification
soilGhg=df_Ghg['Soil Classification'].unique()
print(soilGhg)
for values in soilGhg:
    if(values=='Alfisols' or values=='Fine-loamy, mixed, superactive, mesic Aridic Haplustalfs' 
       or values=='fine-loamy, mixed, superactive, mesic Aridic Haplustalfs'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Planosols,Luvisols,Nitosols')
    if(values=='fine-loamy, mixed, superactive, mesic Ustic Haplocambids'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Cambisols')
    if(values=='Coarse-silty, mixed, mesic Durixerollic Calciorthods'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Vertisols')
    if(values=='Fine-silty, mixed, active, mesic Typic Paleudalfs'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Luvisols,Nitosols')
    if(values=='Fine, mixed, semiactive, mesic Typic Hapludalfs'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Luvisols,Nitosols')
        
    if(values=='frigid Typic Argiustoll'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Acrisols')
    if(values=='Tomek fine, smectitic, mesic Pachic Argiudolls, Filbert fine, smectitic, mesic  Vertic Argialbolls'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Vertisols')
    if(values=='Clayey, illitic, mesic Typic Hapludults'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Acrisols')    
    if(values=='Fine-silty, mixed, superactive, frigid Typic and Pachic Haplustolls'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Kastanozem')
        
        
    if(values=='fine-loamy over sandy or sandy-skeletal, mixed mesic Aridic Argiustoll' 
       or values=='fine-loamy, mixed, mesic Aridic Argiustoll'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Kastanozem')
    if(values=='FINE-LOAMY, MIXED, SUPERACTIVE, FRIGID CALCIC HAPLUDOLLS'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Alfisols')
    if(values=='Fine-silty, mixed, superactive, frigid Typic and Pachic Haplustolls' or 
       values=='Fine-silty, mixed, superactive, frigid Pachic Haplustoll'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Kastanozem')
    if(values=='FINE-LOAMY, MIXED, SUPERACTIVE, FRIGID AERIC CALCIAQUOLLS' or
               values=='FINE-LOAMY, MIXED, SUPERACTIVE, FRIGID TYPIC CALCIAQUOLLS' or 
       values=='FINE-LOAMY, MIXED, SUPERACTIVE, FRIGID CALCIC HAPLUDOLLS'or
           values=='FINE-LOAMY, MIXED, SUPERACTIVE, FRIGID PACHIC ARGIUDOLLS'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Phaeozem')
    if(values=='sandy, mixed, frigid, Entic Haplustoll'or values=='Solonchaks'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Kastanozem')
    if(values=='Clayey, illitic, mesic Typic Hapludults'):
        df_Ghg['Soil Classification']=df_Ghg['Soil Classification'].replace(values,'Luvisols,Nitosols')   
print(df_Ghg['Soil Classification'].unique())
```
