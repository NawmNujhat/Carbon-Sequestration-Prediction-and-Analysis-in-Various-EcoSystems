```python
#print(df_Biomass['Ecosystem'].unique())
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('Temperate Broadleaf forest','Temperate Broadleaf Forest')
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('Tropical/subtropical Forest','Tropical/Subtropical Forest')
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('Upland field ','Upland Field')
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('Tundra ','Tundra')
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('pasture','Pasture')
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('paddy field','Paddy field')
df_Merge['Land-use type']=df_Merge['Land-use type'].replace('Temperate Coniferous forest','Temperate Coniferous Forest')
#print(df_Merge['Land-use type'].unique())
```

```python
unique_columns_LandUseType=df_Merge['Land-use type'].unique()
print(unique_columns_LandUseType)
nan12=[]
for value in unique_columns_LandUseType:
    unique_values_12=df_Merge[df_Merge['Land-use type']==value]['Unnamed: 12'].unique()
    #print(f"Unique values in Column12 associated with '{value}': {unique_values_12.tolist()}")
    if(len(unique_values_12.tolist())==1):
        nan12.append(value)
    else:
        print(value)
print(nan12)
```
